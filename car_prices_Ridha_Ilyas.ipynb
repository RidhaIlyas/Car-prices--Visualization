```python
data=pd.read_csv('car_prices.csv')
data
```

# **About Data**

### Column Description
      1. year:The year in which the car was manufactured.
      2. make: The brand or manufacturer of the car (e.g., Kia, BMW, Volvo).
      3. model: The specific model within the car's make (e.g., Sorento, 3 Series, 560).
      4. trim: The specific version or package of the car model (e.g., LX, 328i, T5).
      5. body: The type of car body (e.g., SUV, Sedan).
      6. transmission: The type of transmission (e.g., automatic, manual).
      7. vin: The Vehicle Identification Number, a unique code identifying each individual vehicle.
      8. state: The state where the car was sold.
      9. condition: The condition of the car at the time of sale (e.g., used, new).
      10. odometer: The mileage on the car at the time of sale.
      11. color: The exterior color of the car.
      12. interior: The interior color of the car.
      13. seller: The seller of the car (e.g., dealership, individual).
      14. mmr: The Manufacturer's Suggested Retail Price (MSRP) of the car.
      15. sellingprice:The actual selling price of the car.
      16. saledate: The date the car was sold.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['sellingprice'],bins=50, edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
```

*   The y-axis represents the frequency or number of cars within each price range, while the x-axis shows the price ranges.

*   The car market is shifted towards lower-priced cars, with a larger
number of cars falling within the lower price ranges.



```python
a=data['sellingprice']>120000
luxury_brands=data[a]
sns.violinplot(x='make',y='sellingprice',data=luxury_brands)
plt.xticks(rotation=90)
plt.title('price distribution across luxury car brands');
```



*   The violin plots visualize the distribution of selling prices for each car make. The thicker parts of the violins indicate higher density of data points, while thinner parts represent lower density.
*   The white dot within each violin plot represents the median selling price for that make



*  Rolls-Royce stands out with the highest median selling price
* Ferrari also has a high median selling price, second only to Rolls-Royce
* Bentley's median price is close to Ferrari’s



```python
data.groupby('year')['sellingprice'].mean().plot(kind='line');
plt.title('trends in car price across years');
```

* This trend could indicate how car prices have evolved over time.





```python
data.groupby(['make','model'])['sellingprice'].max().sort_values(ascending=False).head(20).plot(kind='bar',color='green',edgecolor='black');
plt.title('top 20 luxury models and their makers');
```



*   This chart identifies Ford's Escape is the most expensive or have the highest market value which is followed 458 Italia of Ferrari.
* Among the top 20 luxury cars BMW's M6 Gran Coupe falls on 20th position in terms of selling price




```python
data['model'].value_counts().head(10).plot(kind='pie',autopct='%1.1f%%')
plt.title('Most commonly sold models');
```

* Most commonly sold models are Nissan's Altima followed by Ford's F150
* Out of most common 10 models, Chevrolet's Impala and Grand Caravan shares the last position with 6.9%

```python
data.groupby('color')['mmr'].max().sort_values(ascending=False).head(15).plot(kind='bar',color='silver')
plt.title('correlation between MMR and Color');
```

* this graph reveals that, cars with red color are priced more by the manufacturer than any other colors

```python
data.groupby('transmission')['condition'].mean().plot(kind='bar')
plt.title('how transmission affects the condition of vehicle')
plt.ylabel('condition')
```

* The graph shows no notable difference in condition of automatic and manual vehicles however a condition of automatic vehicles are slightly higher than manual vehicles


```python
b=data[data['year']>2013]
latest_models=b.head(10)
sns.scatterplot(x='year',y='model',data=latest_models)
plt.title('latest models');
```

* the latest models released after 2013 are plotted in the graph with their release year
* most of the vehicles are released in 2014.
* Altima, S60 and Sorento are the only cars that was released in 2015

```python
latest_models.groupby(['make','model'])['sellingprice'].max().sort_values().plot(kind='bar');
plt.title('Latest models, their makers and their prices');
```

* Among the latest models released after 2013, BMW's 6 Series Gran Coupe is of the highest selling price, followed by same make's M5 model
* in the top 10, Chevrolet's Cruze is at the last position
